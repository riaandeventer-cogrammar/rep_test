# EDA Report – Titanic Dataset

This report presents an exploratory data analysis (EDA) of the Titanic dataset. The aim is to identify patterns and key factors that influenced passenger survival.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("Titanic.csv")
```

## Dataset Description

The Titanic dataset contains information about passengers aboard the Titanic, including demographic details, ticket class, fare, and survival status.

Key features include:
- Age: Passenger age
- Sex: Gender of the passenger
- Pclass: Passenger class (1st, 2nd, 3rd)
- Fare: Ticket price
- Survived: Survival status (0 = No, 1 = Yes)

```python
# Fill missing values
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])

# Drop Cabin column
df = df.drop(columns=['Cabin'], errors='ignore')
```

## Data Cleaning

Missing values in the Age column were filled using the median value.  
Missing values in the Embarked column were filled using the mode.  
The Cabin column was dropped due to a high number of missing values.

```python
# Check missing values
df.isnull().sum()
```

```python
sns.heatmap(df.isnull(), cbar=False)
plt.title("Missing Data Heatmap")
plt.show()
```

## Missing Data Analysis

Before cleaning, the dataset contained missing values in the Age and Cabin columns.  
The Age values were imputed, and the Cabin column was removed.  

The heatmap above shows that most missing data has been handled.

```python
sns.countplot(data=df, x='Sex', hue='Survived')
plt.title("Survival by Gender")
plt.show()
```

```python
sns.countplot(data=df, x='Pclass', hue='Survived')
plt.title("Survival by Passenger Class")
plt.show()
```

```python
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

The correlation heatmap shows that passenger class has a negative relationship with survival, meaning that passengers in lower classes were less likely to survive. Fare shows a positive relationship with survival, supporting the idea that wealthier passengers had better survival chances.

```python
sns.boxplot(data=df, x='Survived', y='Fare')
plt.title("Fare Distribution by Survival")
plt.show()
```

## Findings

- Approximately 74% of female passengers survived, compared to around 19% of male passengers.
- First-class passengers had a survival rate of about 63%, while third-class passengers had significantly lower survival rates (~24%).
- Passengers paying higher fares were more likely to survive, indicating a relationship between socio-economic status and survival.
- There is a moderate negative correlation between passenger class and survival, meaning lower-class passengers were less likely to survive.

## Conclusion

The analysis shows that gender and passenger class were the most important factors influencing survival.  

Women and higher-class passengers had a much greater chance of survival, suggesting that social status and evacuation priorities played a major role.  

Overall, the data highlights clear inequalities in survival outcomes during the Titanic disaster.
